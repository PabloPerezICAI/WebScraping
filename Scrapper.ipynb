```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Cargar HTML
url = "https://localhens.com/find-a-farm-results/?post%5B0%5D=farm&address%5B0%5D&distance=50&units=imperial&title&per_page=2000&lat&lng&form=1&action=fs"
response = requests.get(url)

# Parsear HTML
soup = BeautifulSoup(response.content, "html.parser")

farm_divs = soup.find_all('div', class_='content-wrapper')

data = []

for farm_div in farm_divs:
    farm_name = farm_div.find('h2').text.strip()
    location = farm_div.find('div', class_='farm-location').text.strip()
    
    products_div = farm_div.find('div', class_='farm-products')
    if products_div is not None:
        products = products_div.find_all('li')
        product_list = ', '.join([product['title'] for product in products])
    else:
        product_list = ''
    
    where_to_buy_div = farm_div.find('div', class_='farm-where-purchase')
    if where_to_buy_div is not None:
        where_to_buy = where_to_buy_div.find_all('li')
        where_to_buy_list = ', '.join([buy_location['title'] for buy_location in where_to_buy])
    else:
        where_to_buy_list = ''
    
    link_div = farm_div.find('div', class_='farm-link')
    if link_div is not None:
        link = link_div.find('a')['href']
    else:
        link = ''
    
    data.append({
        'Farm': farm_name,
        'Location': location,
        'Products': product_list,
        'Where to Buy': where_to_buy_list,
        'Link': link
    })

df = pd.DataFrame(data)
display(df)
```

```python
data = []
for i in range(0,3338): #los dos primeros
      url = dff['Link'][i]
      response = requests.get(url)
      soup = BeautifulSoup(response.content, "html.parser")
      farm_divs = soup.find_all('main', class_='main')

      for farm_div in farm_divs:
            farm_name = farm_div.find('h1').text.strip()
            adress = farm_div.find('li', class_='farm-address')
            if adress is not None:
                  adress = adress.text.strip()
            else:
                  adress = ''
            phone = farm_div.find('li', class_='farm-phone')
            if phone is not None:
                  phone = phone.text.strip()
            else:
                  phone = ''
            email = farm_div.find('li', class_='farm-email')
            if email is not None:
                  email = email.find('a')['href']
            else:
                  email = ''
            map = farm_div.find('li', class_='farm-directions')
            if map is not None:
                  map = map.find('a')['href']
            else:
                  map = ''
            story = farm_div.find('div', class_='col-12 col-md-8 farm-content-main entry-content')
            if story is not None:
                  story = story.text.strip()
            else:
                  story = ''
            hours = farm_div.find('li', class_='farm-additional-hours')
            if hours is not None:
                  hours = hours.text.strip()
            else:
                  hours = ''
            
      data.append({
            'Farm': farm_name,
            'Adress': adress,
            'Phone': phone,
            'Email': email,
            'Map': map,
            'Story': story,
            'hours': hours
            })
df2 = pd.DataFrame(data)
display(df2)
```
